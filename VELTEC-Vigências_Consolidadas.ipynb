## Análise de Predição - Pontuação do Motorista

A pontuação final do motorista por vigência deve ser considerada através da soma de todas as pontuações adquiridas na vigência.

```python
import pandas as pd
```

```python
df=pd.read_csv('vigencias_consolidadas.csv', quotechar="'")
```

```python
df.head()
```

```python
# Selecionar colunas específicas
cols_pont = ['pontos_diagnostico_obd',
             'pontos_distracao', 'pontos_fadiga', 'pontos_manuseio_celular',
             'pontos_possivel_colisao',
             'pontos_previsao_colisao', 'pontos_remocao_equipamento',
             'pontos_cigarro', 'pontos_sem_vigencia',
             'pontos_curvas', 'pontos_distancia_perigosa',
             'pontos_aceleracoes', 'pontos_frenagens', 'pontos_trepidacoes',
             'pontos_banguela', 'pontos_motor_ocioso', 'pontos_rpm_faixa_1',
             'pontos_vel_faixa_1', 'pontos_vel_faixa_2', 'pontos_vel_faixa_3',
             'pontos_vel_via_faixa_1', 'pontos_vel_via_faixa_2',
             'pontos_vel_via_faixa_3','pontos_camera_obstruida',
             'pontos_motorista_ausente','pontos_sem_cinto']
```

```python
df['pontos_final'] = df[cols_pont].sum(axis=1)
```
